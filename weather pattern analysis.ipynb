```python
import pandas as pd
import matplotlib.pyplot as plt
from datetime import datetime
import seaborn as sns

df=pd.read_csv("weather.csv")

print(df)
print()
print(df.info())
print()
print(df.describe())
```

```python
df["DATE"]=pd.to_datetime(df["DATE"])
```

```python
def search(city):
    return df[df["CITY"] == city]



def temp_trends(data):
    plt.figure(figsize=(10,6))
    plt.plot(data["DATE"], data["TEMP"], color="red", label="Temperature")
    plt.xlabel("Date")
    plt.ylabel("Temperature (C)")
    plt.title("Temperature trends in certain city")
    plt.grid(True)
    plt.legend()
    plt.show()



```

```python
def precip_trends(data):
    plt.figure(figsize=(10,6))
    plt.plot(data["DATE"], data["PRECIP"], color="blue", label="Precipitation")
    plt.xlabel("Date")
    plt.ylabel("Precipitation")
    plt.title("Precipitation trends in certain city")
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
def humid_trends(data):
    plt.figure(figsize=(10,6))
    plt.plot(data["DATE"], data["HUMIDITY"], color="green", label="Humidity")
    plt.xlabel("Date")
    plt.ylabel("Humidity")
    plt.title("Humidity trends in certain city")
    plt.grid(True)
    plt.legend()
    plt.show()
```

```python
city=input("Choose City to check")
city_data=search(city)

if city_data.empty:
    print("No data foun for this city")
else:

    temp_trends(city_data)
    precip_trends(city_data)
    humid_trends(city_data)
```
